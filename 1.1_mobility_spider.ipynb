```python
from bs4 import BeautifulSoup
from bs4 import UnicodeDammit
import requests as req
import re
import json
import pandas as pd
import numpy as np

def Get(date,id):
    headers={
    
    'Accept' : 'text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3;q=0.7',
    'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/117.0.0.0 Safari/537.36 Edg/117.0.2045.36'
    }
    url="https://huiyan.baidu.com/migration/cityrank.jsonp?dt=city&id=%s&type=move_in&date=%s&callback=json"%(date,id)
    print(url)
    r=req.get(url,headers=headers)
    txt=r.text
    reg=r"(\\u[a-z0-9]{4}){1,}"
    m=re.search(reg,txt)
    while m!=None:
        start=m.start()
        end=m.end()
        name=txt[start:end]
        city=txt[start:end].encode('utf-8').decode('unicode_escape')
        txt=txt.replace(name,city)
        m=re.search(reg,txt)
    js=json.loads(txt.replace("\njson(","").replace(")",""))
    return js
#def getID(path):
    ids=pd.read_csv(path)
    return ids.iloc[:,1]
def getSumIntensity(path):
    ins=pd.read_csv(path,index_col=0)
    return ins
def outTable(path):
    names=pd.read_csv(path,index_col=0)
    names=names.fillna(0)
    return names


id_path=r"C:\Users\86157\Desktop\2023summer\baidu_mobility\id.csv"
SumIn_path=r"C:\Users\86157\Desktop\id_sum_in.csv"
out_path=r"C:\Users\86157\Desktop\20230405.csv"
dates=['20230406','20230407']
#ids=getID(id_path)
ins=getSumIntensity(SumIn_path)
names=outTable(out_path)
ids=names.index
cols=names.columns
data=[]
row={}
for i in dates:
    for j in ids:
        intensity=float(ins.loc[int(j),i])
        js=Get(str(j),i)
        cities=js['data']['list']
        for city in cities:
            city_name=city['city_name']
            #print(city)
            #print(city['value'])
            item=intensity*float(city['value'])*0.01
            #print(names.loc[j,city])
            if city_name in cols:
                names.loc[j,city_name]=names.loc[j,city_name]+item
    names.to_csv("C://Users/86157/Desktop/%s.csv"%i)

        



```
